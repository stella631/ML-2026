# 아래 코드 작성 시 발생할 수 있는 에러
1. Mol이 None인 경우 (분자 구조로 인식하지 못하는 경우가 있는지 확인)
2. Descriptor값이 확인되지 않음 (계산 실패. 구조가 불명확해서)

```python
신혜지 = []
```

```python
#Quiz 1 (표현자 계산해보기)
    #SMILES 코드 저장 (row는 엑셀에 있는 한줄을 의미)
    #RDKit Mol로 저장
    #Descriptor 계산
    #Descriptor 정보를 '신혜지'에 append

# =========================================
# Quiz 1
# 전체 물질에 대해 RDKit Descriptor 계산
# =========================================

import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# csv 불러오기
df = pd.read_csv('skin_irritation_cleaned.csv')

# descriptor 저장용 리스트
descriptor_list = []

# 전체 분자 반복
for index, row in df.iterrows():
    
    # smiles 저장
    smiles = row['SMILES']
    
    # Mol 변환
    mol = Chem.MolFromSmiles(smiles)
    
    # Mol 생성 실패 시 제외
    if mol is None:
        continue
    
    try:
        # descriptor 계산
        descriptors = Descriptors.CalcMolDescriptors(mol)
        
        # 추가 정보 저장
        descriptors['label'] = row['label']
        descriptors['Chemical_Name'] = row['Chemical_Name']
        descriptors['SMILES'] = smiles
        
        # 리스트에 추가
        descriptor_list.append(descriptors)
    
    except:
        continue

# dataframe 변환
descriptor_df = pd.DataFrame(descriptor_list)

print(descriptor_df.shape)

descriptor_df.head()
```

Last Quiz: dataframe에서 단일 구조만 남기고 모두 제외하면 데이터가 몇개 남을까요?

```python
# =========================================
# Last Quiz
# 단일 구조만 남기기
# '.' 포함된 경우 제거
# =========================================

df_single = df[
    ~df['SMILES'].str.contains(r'\.', regex=True)
]

print(df.shape)
print(df_single.shape)

print(df_single['label'].value_counts())
```
